# Description of the repartition of data

```python
df.rating = pd.to_numeric(df.rating)
df.dtypes
```

```python
ratings = pd.DataFrame(df.groupby('jokeid')['rating'].mean())
display(ratings.head(5))
ratings.shape
```

```python
user_ratings = pd.DataFrame(df.groupby('Userid')['rating'].mean())
display(user_ratings.head(5))
display(user_ratings.shape)
```

```python
user_jokes = pd.DataFrame(df.groupby('Userid')['jokeid'].count())
display(user_jokes.head(5))
display(user_jokes.shape)
```

```python
user_rate = pd.DataFrame(df.groupby('jokeid')['Userid'].count())
display(user_rate.head(5))
display(user_rate.shape)
```

```python
fig = plt.figure(figsize=(13, 13))
        
plt.subplot2grid((2,2), (0, 0))
ratings['rating'].hist(bins=50)
plt.legend()
plt.xlabel('Average mean')
plt.ylabel('Number of jokes')
plt.title("Repartition of the average mean by jokes")


plt.subplot2grid((2,2), (0, 1))
user_ratings['rating'].hist(bins=50)
plt.legend()
plt.xlabel('Average mean')
plt.ylabel('user')
plt.title("Repartition of the average mean rate by user")
        
plt.subplot2grid((2,2), (1, 0))
user_jokes['jokeid'].hist(bins=50)
plt.legend()
plt.xlabel('Jokes')
plt.ylabel('Number of users who rates a joke')
plt.title("Repartition of number of rating from a user by jokes")

plt.subplot2grid((2,2), (1, 1))
user_rate['Userid'].hist(bins=50)
plt.legend()
plt.xlabel('Number of jokes rated')
plt.ylabel('User')
plt.title("Repartition of number of jokes rated by user")

plt.show()
```

```python
new['number_of_ratings'] = pd.DataFrame(df.groupby('jokeid')['rating'].count())
display(new.shape)
display(new.dtypes)
display(new.tail(5))
sns.jointplot(x=ratings['rating'], y=new['number_of_ratings'])
```

```python
#display the jokes with the most rates
new.sort_values('number_of_ratings', ascending=False).head(10)
```

# Creation of the matrix used for the recommendation system

```python
joke_matrix = df.pivot_table(index='Userid', columns='jokeid', values='rating', fill_value=0)
display(joke_matrix.shape)
display(joke_matrix.head(5))
display(joke_matrix.tail(5))
```

```python
test_best_joke = joke_matrix[7]
test_seconf_joke = joke_matrix[8]
display(test_best_joke.shape)
display(test_best_joke.head())
display(test_seconf_joke.shape)
display(test_seconf_joke.head())
```

```python
similar_to_joke7 = movie_matrix.corrwith(test_best_joke)
similar_to_joke8 = movie_matrix.corrwith(test_seconf_joke)
display(similar_to_joke7.head())
display(similar_to_joke8.head())
```

```python
similar_to_joke7 = similar_to_joke7.join(new['number_of_ratings'])
similar_to_joke8 = similar_to_joke8.join(new['number_of_ratings'])
display(similar_to_joke7.head())
display(similar_to_joke8.head())
```
