# Description 

Our dataset has no null value

Over 1.7 million continuous ratings (-10.00 to +10.00) of 150 jokes from 59,132 users: collected between November 2006 - May 2009

Save to disk, then unzip: jester_dataset_2.zip (7.7MB)

Format:

    jester_ratings.dat: Each row is formatted as [User ID] [Item ID] [Rating]
    jester_items.dat: Maps item ID's to jokes

Note that the ratings are real values ranging from -10.00 to +10.00. As of May 2009, the jokes {7, 8, 13, 15, 16, 17, 18, 19} are the "gauge set" (as discussed in the Eigentaste paper) and the jokes {1, 2, 3, 4, 5, 6, 9, 10, 11, 12, 14, 20, 27, 31, 43, 51, 52, 61, 73, 80, 100, 116} have been removed (i.e. they are never displayed or rated). 

 Format:

    In this dataset we stripped out users that did not respond to the gauge set of question. The data is formated as an excel file representing a 66336 x 151 matrix with rows as users and columns as jokes.
    10 of the jokes don't have ratings, their ids are: { 1, 2, 3, 4, 6, 9, 10, 11, 12, 14 }.
    Each rating is from (-10.00 to +10.00) and 99 corresponds to a null rating (user did not rate that joke).

# Find the real mean value for each column

```python
cpt_col_good = 0
cpt_col_wrong = 0
test_value = 0.5
col_mean = []
for col in data.columns:
    tot = 0 #used to add all the rated values
    cpt = 0 # number of data with rated values
    for val in data[col]:
        if (val > -10 and val <10):
            tot +=val
            cpt +=1
    if cpt != 0:
        mean_col = tot / cpt
    else :
        mean_col = 0
    print ("Mean score for ", col, " : ", mean_col)
    col_mean.append(mean_col)
    if (mean_col > test_value):
        cpt_col_good +=1
    if (mean_col < -test_value):
        cpt_col_wrong +=1
```

```python
print("number of columns a mean score > +%s : "%test_value, cpt_col_good)
print("number of columns a mean score < -%s : "%test_value, cpt_col_wrong)
```

```python
print (col_mean)
```

```python
fig = plt.figure(figsize=(8, 8))
plt.plot(col_mean, label='Average score', color='chocolate')
plt.legend()
plt.xlabel('Column')
plt.ylabel('Average score')
plt.title("Average score by column")
plt.show()
```

```python
index = []
couleurs = []

for i in range(len(col_mean)):
    index.append(i)
    if (col_mean[i] > 0 and col_mean[i] <=1): couleurs.append("r")
    else :
        if (col_mean[i] < 0 and col_mean[i] >=-1): couleurs.append("y")
        else :
            if (col_mean[i] > 1 and col_mean[i] <2) : couleurs.append("c")
            else : 
                if (col_mean[i] < -1 and col_mean[i] > -2) : couleurs.append("g") 
                else : couleurs.append("b")

fig = plt.figure(figsize=(8, 8))
plt.scatter(index,col_mean,c=couleurs)
plt.legend()
plt.xlabel('Column')
plt.ylabel('Average score')
plt.title("Average score by column")
plt.show()
```

# Find the real mean value for each row

```python
cpt_row_good = 0
cpt_row_wrong = 0
test_value_row = 1
row_mean = []
for row in range(len(data)):
    tot = 0 #used to add all the rated values
    cpt = 0 # number of data with rated values
    for i in data.iloc[row]:
        if (i > -10 and i <10):
            tot +=i
            cpt +=1
    if cpt != 0:
        mean_row = tot / cpt
    else :
        mean_row = 0
    print ("Mean score for ", row, " : ", mean_row)
    row_mean.append(mean_row)
    if (mean_row > test_value_row):
        cpt_row_good +=1
    if (mean_row < -test_value_row):
        cpt_row_wrong +=1
```

```python
print("number of rows a mean score > +%s : "%test_value_row, cpt_row_good)
print("number of rows a mean score < -%s : "%test_value_row, cpt_row_wrong)
```
